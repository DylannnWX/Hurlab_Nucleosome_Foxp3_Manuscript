```python
import pandas as pd
import numpy as np
```

```python
OCR_Fimo=pd.read_csv("TnG_Fimo.tsv", sep='\t', skipfooter=3,  engine='python')
```

```python
OCR_Fimo.head(5)
```

```python
values_to_keep = ['12-AAAAAACAAAAAACAAAAAAMAAAAAA', 
                  '24-AAAACAAAAACAAAA', 
                  '1-AAACAAACAAACAAACA',
                  '13-AACAACAACAACAAMA', 
                  '7-AAAACAAAAACAAAAACAAAAAM', 
                  '4-AAAACAAAACAAAACAAAAC',
                  '14-AAACAAAAYAAAACA']

# Assuming df is your DataFrame
OCR_Fimo_TnG = OCR_Fimo[OCR_Fimo['motif_id'].isin(values_to_keep)]
```

```python
OCR_Fimo_TnG.shape, OCR_Fimo.shape
```

```python
strings_to_remove = ["aaaaaaaaaaaa", "acacacacacac", "tttttttttttt", "tgtgtgtgtgtg", "agagagagagag", "gcgcgcgcgcgc","atatatatatat","tctctctctctc"]
```

```python
strings_to_remove_lower = [s.lower() for s in strings_to_remove]
```

```python
pattern = '|'.join(strings_to_remove_lower)

# Filtering out rows where any substring contains the specified strings
OCR_Fimo_TnG = OCR_Fimo_TnG[~OCR_Fimo_TnG['matched_sequence'].str.lower().str.contains(pattern)]
```

```python
OCR_Fimo_TnG.shape
```

```python
OCR_Fimo_TnG2 = OCR_Fimo_TnG[OCR_Fimo_TnG['p-value'] < 8e-5].copy()
```

```python
OCR_Fimo_TnG2.shape
```

```python
OCR_Fimo_TnG2.head(5)
```

```python
OCR_Fimo_TnG2[['chr', 'numberA-numberB']] = OCR_Fimo_TnG2['sequence_name'].str.split(':', expand=True)
OCR_Fimo_TnG2[['motif_start', 'motif_end']] = OCR_Fimo_TnG2['numberA-numberB'].str.split('-', expand=True)
```

```python
OCR_Fimo_TnG2.head(5)
```

```python
OCR_Fimo_TnG2['TnG_start']=OCR_Fimo_TnG2['motif_start'].astype(int) + OCR_Fimo_TnG2['start'].astype(int)-1
OCR_Fimo_TnG2['TnG_end']=OCR_Fimo_TnG2['motif_start'].astype(int) + OCR_Fimo_TnG2['stop'].astype(int)
```

```python
OCR_Fimo_TnG2.head(5)
```

```python
OCR_TnG_coords = OCR_Fimo_TnG2[['chr', 'TnG_start', 'TnG_end']].copy()
OCR_TnG_coords.to_csv('nG_coords.bed', sep ='\t', header=False,index=False)
```
